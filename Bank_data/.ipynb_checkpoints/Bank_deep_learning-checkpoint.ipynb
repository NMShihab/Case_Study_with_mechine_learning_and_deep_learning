# Implementation of deep learning 

```python
# Import packages
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from sklearn.metrics import confusion_matrix, accuracy_score,classification_report
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
from scipy.stats import randint
from sklearn.model_selection import cross_val_score , GridSearchCV, RandomizedSearchCV
%matplotlib inline
```

```python
# Import data
data = pd.read_csv('Churn_Modelling.csv')
```

```python
data.head()
```

```python
feature = data.iloc[:,3:13].values # Feature for training 
target = data.iloc[:,13].values # Target
```

```python
feature
```

**Encoding catagorical data**

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder     # Import package for encoding
labelencoder_X_1 = LabelEncoder()                                 # Create object of LabelEncoder
feature[:, 1] = labelencoder_X_1.fit_transform(feature[:, 1])     # Encode 'Geography' column
labelencoder_X_2 = LabelEncoder()                                 # Create anoher object of LabelEncoder
feature[:, 2] = labelencoder_X_2.fit_transform(feature[:, 2])     # Encode 'Gender' column
onehotencoder = OneHotEncoder(categorical_features = [1])         # Create OneHotencoder object for 1 feature
feature = onehotencoder.fit_transform(feature).toarray()          # Create dummy variables 
feature = feature[:, 1:]                                          # Remove one column to avoid dummy variable trap
```

```python
feature.shape
```

**Splitting the dataset into the Training set and Test set**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(feature, target, test_size = 0.2, random_state = 0)
```

**Feature Scaling**

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# Importing the Keras libraries and packages
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
# Initialising the ANN
classifier = Sequential()
```

```python
# Adding the input layer and the first hidden layer
classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu', input_dim = 11))
```

```python
# Adding the second hidden layer
classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu'))
```

```python
# Adding the output layer
classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
```

```python
# Compiling the ANN
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
# Fitting the ANN to the Training set
classifier.fit(X_train, y_train, batch_size = 10, epochs = 100)
```

```python
# Predicting the Test set results
y_pred = classifier.predict(X_test)
```

```python
y_pred = (y_pred > 0.5)
```

```python
y_pred
```

**Evaluation of Model**

```python
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score
```

```python
def build_classifier():
    classifier = Sequential()
    classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu', input_dim = 11))
    classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu'))
    classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
    classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
    return classifier
```

```python
classifier = KerasClassifier(build_fn = build_classifier, batch_size = 10, epochs = 100)
```

```python
accuracies = cross_val_score(estimator = classifier, X = X_train, y = y_train, cv = 10, n_jobs = -1)
```

```python
mean_accuracy = accuracies.mean()
```

```python
variance_accuracy = accuracies.std()
```
